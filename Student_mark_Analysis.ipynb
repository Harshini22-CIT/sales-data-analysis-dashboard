```python
import csv

filename = "student_marks.csv"

total_students = 0
pass_count = 0
fail_count = 0

total_marks = 0
highest_marks = 0
top_student = ""

print("=" * 50)
print("STUDENT MARKS ANALYSIS")
print("=" * 50)

with open(filename, "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        total_students += 1

        marks = int(row["FinalExamScore"])
        total_marks += marks

        if marks > highest_marks:
            highest_marks = marks
            top_student = row["Name"]

        if row["PredictedResult"] == "Pass":
            pass_count += 1
        else:
            fail_count += 1

average_marks = total_marks / total_students

print(f"Total Students      : {total_students}")
print(f"Passed Students     : {pass_count}")
print(f"Failed Students     : {fail_count}")
print(f"Average Marks       : {average_marks:.2f}")
print(f"Highest Marks       : {highest_marks}")
print(f"Top Student         : {top_student}")

print("\nStudents likely to Fail:")

with open(filename, "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        if row["PredictedResult"] == "Fail":
            print(f"{row['Name']} - Attendance: {row['Attendance']}%, Final Marks: {row['FinalExamScore']}")
```
