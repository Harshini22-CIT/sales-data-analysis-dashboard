```python
import csv

filename = "credit_card_transactions.csv"

total_transactions = 0
fraud_transactions = 0
legitimate_transactions = 0
total_amount = 0
fraud_amount = 0

fraud_by_location = {}

with open(filename, "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        total_transactions += 1

        amount = float(row["Amount"])
        total_amount += amount

        if row["Status"] == "Fraud":
            fraud_transactions += 1
            fraud_amount += amount

            location = row["Location"]
            fraud_by_location[location] = fraud_by_location.get(location, 0) + 1
        else:
            legitimate_transactions += 1

print("=" * 50)
print(" CREDIT CARD FRAUD ANALYSIS ")
print("=" * 50)

print(f"Total Transactions      : {total_transactions}")
print(f"Legitimate Transactions : {legitimate_transactions}")
print(f"Fraud Transactions      : {fraud_transactions}")
print(f"Total Transaction Amount: ₹{total_amount:,.2f}")
print(f"Fraud Amount            : ₹{fraud_amount:,.2f}")

fraud_rate = (fraud_transactions / total_transactions) * 100
print(f"Fraud Rate              : {fraud_rate:.2f}%")

print("\nFraud Cases by Location")
print("-" * 30)

for location, count in fraud_by_location.items():
    print(f"{location:<15} {count}")
```
